```python
# finds all the possible neigbhors of the coordinate given returns a list of coordinates
# coordinate: a list of the x, y coordinate
# main_map: the main map where the coordinate is from
def find_neighbors(coordinate, main_map):
    width = len(main_map[0])
    height = len(main_map)
    
    neighbors =[]
    
    if coordinate[0] > 0:
        neighbors.append([coordinate[0] - 1, coordinate[1]])
    if coordinate[0] < width - 1:
        neighbors.append([coordinate[0] + 1, coordinate[1]])
    if coordinate[1] > 0:
        neighbors.append([coordinate[0] , coordinate[1]- 1])
    if coordinate[1] < width - 1:
        neighbors.append([coordinate[0] , coordinate[1]+ 1])
        
    return neighbors
    
    
```

```python
# inserts a qeue into a priority qeue
# qeue: a list of values in the priority qeue 
# new_value: the new node being added to the qeue
def insert_priority(qeue, new_value):
    qeue.append(new_value)
    aranged = False
    
    index = len(qeue) - 1 
    
    while index > 0 and aranged == False:
        if qeue[index][2] < qeue[index - 1][2]:
            temp = qeue[index]
            qeue[index] = qeue[index - 1]
            qeue[index - 1] = temp
        else:
            aranged = True
        
        index -= 1
    
```

```python
# checks where the coordinate index is at in the list returns -1 if it is not there
# qeue:
def find(qeue, coordinate):
    index = -1
    for i in range(len(qeue)):
        if qeue[i][0] == coordinate:
            index = i
            
    return index
```

```python
def check_and_add(qeue_, finished_, every_, new_addition_):
    coordinate = new_addition_[0]
    in_finished = find(finished_, coordinate)
    
    if in_finished == -1:
        in_qeue = find(qeue_, coordinate)
        
        if in_qeue == -1:
            insert_priority(qeue_, new_addition_)
            every_.append(new_addition_)
        else:
            new_value = new_addition_[2]
            old_value = qeue_[in_qeue][2]
            if new_value < old_value:
                del qeue_[in_qeue]
                insert_priority(qeue_, new_addition_)
                every_.append(new_addition_)
            
    
```

```python
# calculets the distance from the current sit to the end
# current: a coordinate of the start
# end: a coordinate of the end
def dis_calculator(current, end):
    x1 = current[0]
    x2 = end[0]
    y1 = current[1]
    y2 = end[1]
    
    x = 0
    y = 0
    
    if x1 > x2:
        x = x1 - x2
    else:
        x = x2 -x1
    
    if y1 > y2:
        y = y1 - y2
    else:
        y = y2 - y1
        
    total = (x*x) + (y*y)
    
    distance = total**(.5)
    
    return distance
    
```
